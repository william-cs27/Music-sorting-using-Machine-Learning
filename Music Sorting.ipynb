```python
##Top Songs of the year 2018 in Spotify
##Different patterns in audio features of the songs
##

import pandas
import numpy
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
#marking a list of missing values

missing_values = ["--","na","__","","NaN"]
```

```python
df = pandas.read_csv(r"Downloads\New folder\top2018.csv", na_values = missing_values)
df.head(100)
```

```python
print("Total missing feature wise seperated",'\n')
print("Out:")
print(df.isnull().sum())
print("\n")
```

```python
##Top 10 artists with most songs

df['artists'].value_counts().head(10)
```

```python
##Danceability of Songs

fig, ax1 = plt.subplots(figsize =(200,10))
sns.barplot(x='name', y='danceability', data = df)
plt.title("Danceability of Songs")
plt.show()
```

```python
##Top 10 Songs wrt Danceability

df[['name','artists','danceability','tempo']].sort_values(by='danceability',ascending=False).head(10)
```

```python
##Energy of Songs

fig, ax1 = plt.subplots(figsize =(200,10))
sns.barplot(x='name', y='energy', data = df)
plt.title("Energy of Songs")
plt.show()
```

```python
##Top 10 Songs wrt to Energy

df[['name','artists','energy','tempo']].sort_values(by='energy',ascending=False).head(10)
```

```python
##Tempo of Songs

fig, ax1 = plt.subplots(figsize =(200,10))
sns.barplot(x='name', y='tempo', data = df)
plt.title("Tempo of Songs")
plt.show()
```
